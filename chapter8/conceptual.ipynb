2. A stump is the tree with depth of one. Follow 8.12 we have  
\begin{align}
\hat f (x) &= \sum_{b=1}^{B} \lambda \hat f^b(x) \\ 
\hat f (x) &= \lambda \hat f^1(x) + \lambda \hat f^2(x) +...+\lambda \hat f^B(x) \\
\hat r_i &= y_i - \lambda\hat f^1(x_i) - \lambda\hat f^2(x_i)-...-\lambda\hat f^B(x_i)
\end{align}

Since we split using a single point, after every b in B we get 2 splits. After
B splits, by adding them up, we just divide the data using multiples of points.
Thus, the model is broken up into multiple parts or f(xi). Or   
\begin{align}
f(X) = \sum_{j = 1}^p f_j(X_j)
\end{align}

3. We have p_m_1 = 1 - p_m_2

```python
# Parameters
p = np.arange(0, 1.0, 0.01)

# classification error
E = [1 - x for x in np.maximum(p,1-p)]

# Gini index
G = [x*(1-x) + x*(1-x) for x in p]

# cross-entropy
D = [-x*np.log(x) - (1-x)*np.log(1-x) for x in p]

# Plot
plt.plot(p, E, label='Classification Error', c='g')
plt.plot(p, G, label='Gini Index', c='r')
plt.plot(p, D, label='Cross-entropy', c='b')
plt.legend()
```
