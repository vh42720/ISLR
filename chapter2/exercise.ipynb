5. Advantages of a very flexible approach is a much closer fit (especially for non linear) overall and
decrease the bias. Disadvantages of a very flexible approach are over fitting data (get too much noise),
increasing variance and hard to get inferences.  
For a non linear relationship, it is generally better to use flexible approach. Moreover, if you want
predictions instead of inferences, flexible approach will also perform better.
for a linear relationship, inflexible approach is preferred. It also gives you better interpretability.

6. A parametric approaches assume the response takes the form of function f which is dependent upon
p predictors. The advantage of a parametric approach is the smaller number of predictors we have to 
keep track and a smaller sample size needed. The disadvantage of a parametric approach is either 
it is too simple or we over fit the data.  
A non-parametric approaches does not assume any form and thus takes a large amount of data to create 
estimate f.
